```python
#10 columna date y cantidad choques

df_moto['dia'] = df_moto['crash_date'].dt.day
df_moto['parte_mes'] = df_moto['dia'].apply(lambda x: 'Principio de mes' if x <= 15 else 'Fin de mes')

choques_parte_mes = (
    df_moto.groupby('parte_mes')['collision_id']
    .count()
    .reset_index(name='cantidad_choques')
)
print("\n1) Choques al principio o fin de mes:")
print(choques_parte_mes)
```

```python
#11 calles con mas choques

# 3) Calles principales (on_street_name) vs choques
# ==============================================
choques_calle = (
    df_moto['on_street_name']
    .value_counts()
    .reset_index()
)
choques_calle.columns = ['on_street_name', 'cantidad_choques']
print("\n3) Calles con m치s choques:")
print(choques_calle.head(10))
```

```python
#12 comparar cruces con choques con respecto a la cantidad de choques en cual de estos es mayor 

choques_cruce = (
    df_moto['cross_street_name']
    .value_counts()
    .reset_index()
)
choques_cruce.columns = ['cross_street_name', 'cantidad_choques']
print("\n4) Cruces con m치s choques:")
print(choques_cruce.head(10))
```

```python
#13tipo de vehiculo con respecto a choques

vehiculos = pd.concat([df_moto['vehicle_type_code_1'], df_moto['vehicle_type_code_2']])
choques_vehiculo = (
    vehiculos.value_counts()
    .reset_index()
)
choques_vehiculo.columns = ['tipo_vehiculo', 'cantidad_choques']
print("\n5) Tipos de veh칤culos con m치s choques:")
print(choques_vehiculo.head(10))
```

```python
#14 cruzar con dataset de calendario para verificar con columna date si es dia de fin de semana y comparar con la cantidad de choques

# Asegurar formato de fecha correcto
df_moto['crash_date'] = pd.to_datetime(df_moto['crash_date'], errors='coerce')

# Crear una columna que indique si el d칤a es fin de semana
# weekday(): lunes=0 ... domingo=6
df_moto['es_fin_de_semana'] = df_moto['crash_date'].dt.weekday >= 5

# Agrupar por esa columna
choques_semana = (
    df_moto.groupby('es_fin_de_semana')['collision_id']
    .count()
    .reset_index(name='cantidad_choques')
)

# Reemplazar True/False por texto m치s legible
choques_semana['tipo_dia'] = choques_semana['es_fin_de_semana'].map({
    True: 'Fin de semana (S치b/Dom)',
    False: 'D칤a h치bil (Lun-Vie)'
})
choques_semana = choques_semana[['tipo_dia', 'cantidad_choques']]

# Mostrar resultados
print("\n9) Comparaci칩n de choques entre d칤as h치biles y fines de semana:")
print(choques_semana)
```

```python
#15 cual es el porcentaje de victimas fatales y lesionadas

total_lesionadas = df_moto['number_of_persons_injured'].sum()
total_fatales = df_moto['number_of_persons_killed'].sum()
total_victimas = total_lesionadas + total_fatales

porc_lesionadas = (total_lesionadas / total_victimas) * 100 if total_victimas > 0 else 0
porc_fatales = (total_fatales / total_victimas) * 100 if total_victimas > 0 else 0

print("\n7) Porcentaje de v칤ctimas:")
print(f"Lesionadas: {porc_lesionadas:.2f}%")
print(f"Fatales: {porc_fatales:.2f}%")
```
